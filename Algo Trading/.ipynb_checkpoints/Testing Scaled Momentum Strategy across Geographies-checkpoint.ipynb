## 2. Data Modelling

return_month_trading_days will add an extra column to the input in which will be detailed the total amount of trading days for each month. Going forward, this will allow us to compute strictly correct variance forecast and consequently scaled or risk-managed momentum returns.

```python
def return_month_trading_days(x):
    trading_days = np.zeros(len(x))
    month_temp = 0 
    temp = 0 
    counter_month_day = 1
    for idx, values in enumerate(x.TimePeriod):
        if values.month != month_temp:
            trading_days[temp:idx] = counter_month_day
            month_temp = values.month
            counter_month_day = 1
            temp = idx
        elif values.month == month_temp:
            counter_month_day +=1
            #helper for last months given function construction
            if ((values.year == 2020) and (values.day == 30) and (values.month == 4)):
                trading_days[temp:idx+1] = counter_month_day
    x["BusinessDays"] = trading_days
    return x
```

The following function computes the variance forecast by looking at the past 6 months returns following a strict day counting framework. One might argue that through the sum and division terms of the formula, strictly counting the days will not make any difference due to the natural elimination of both terms. Nonetheless, it can make a substantial difference once we convert the daily realized variance into a monthly realized variance. Indeed, prior to 1952, trading days per week were equal to six as Saturday were open to trading, hence almost one fifth of the sample would be incorrectly estimated using 21 trading day as a general rule. That would lead to a systematic under estimation of the realized variance for this part of the sample.

```python
def realized_variance_test(data):
    data = return_month_trading_days(data)
    realized_variance = np.zeros(len(data))
    realized_var_6months = dict()
    month_counter = 0
    month_temp = 0
    month_day_sum = [] #will allow to count trading days for the past 6 months
    realized_variance = data["Mom"]**2
    for idx, values in enumerate(data.TimePeriod):
        if values.month != month_temp:
            month_temp = values.month
            if month_counter < 6:
                month_counter +=1
                month_day_sum.append(int(data.BusinessDays.iloc[idx-1]))
            else:
                month_day_sum.pop(0)
                month_day_sum.append(int(data.BusinessDays.iloc[idx-1]))
                temp = sum(month_day_sum)
                for i in range(int(data["BusinessDays"].iloc[idx])):
                    #print(temp) #uncomment here to check the amount of trading days that really happened during the past 6 months
                    realized_var_6months[(data["TimePeriod"].iloc[idx+i])] = ((sum(realized_variance[(idx-temp):idx])/temp)*data["BusinessDays"].iloc[idx])
    return realized_var_6months
                
```

Finally, we perform some modelling to the dataframe so as to get the returns for the scaled momentum strategy and be able to compare its risk/return characteristics with those of a vanilla momentum strategy.

```python
def modelling(realized_var, data):
    #Create a Merged entity that takes the data excel and includes the realized variance
    df = pd.DataFrame.from_dict(realized_var, orient="index")
    last = pd.merge(data, df, how = 'left', left_on = ['TimePeriod'] , right_on = df.index)
    
    #Cleanse and label the table
    last = last.rename(columns = {0: 'Realized Variance'})
    last = last.fillna(method = "ffill").dropna()
    
    #Construct Scaled momemtum returns
    sigma_target = 12 * (np.sqrt(1/12)) #we transform the annual standard variation given in the paper into a monthly standard variation
    last["Scaling Factor"] = sigma_target / (np.sqrt(last["Realized Variance"]))
    last["Scaled Mom"] = last["Scaling Factor"] * last["Mom"]
    
    #Construct cumulative returns
    cum_mom = (1+(last["Mom"]/100)).cumprod()
    cum_scaled_mom = (1+(last["Scaled Mom"])/100).cumprod()
    last["cum_mom"] = cum_mom
    last["Scaled Cum Mom"] = cum_scaled_mom
    last = last.drop(columns = ["BusinessDays"])
    
    return last
```

```python
df = modelling(realized_variance_test(data), data)
```

## 3. Results Interpretations

### 1. The US

As can be seen below, our implementation of the paper differed in regard to the scaling factor. Quite frankly, it is hard to understand why as the realized variance did not show some extreme deviation versus the paper, even after opting to use a very strict day counting framework. This ultimately cause some problems in terms of practical implementations as achieving such a scaling factor would demand substantial leverage. Indeed, even if possible, this would result in substantial costs impacting the returns. Hence, it is important to keep in mind the latter while studying the performance of the strategy.

In any case, as can be seen below, the scaled strategy greatly outperforms the normal momentum strategy with a Sharpe ratio of 1.08 vs 0.56. Furthermore, the ultimate goal of such strategy being to reduce the drawdowns, we can see that it is a success. Indeed, the maximum drawdowns were reduced from -69.8% to -47.3%. Actually, even though the returns still exhibit some heavy tails, they are significantly reduced from 26.43 to 10.79. The skewness is also halved from -1.59 to -0.87.
To conclude, even though the results of our implementation of the scaled momentum strategy may not be exact, we still reach the same goal of the authors of the paper in greatly reducing the huge drawdowns that come with the momentum strategy for the US geography. We will expand this study to Europe, Japan and Asia Pacific Ex-Japan below.

### 2. Europe, Japan and Asia Pacific Ex-Japan

First we construct the dataframe with the needed returns

```python
dic = {"Europe_MOM_Factor_Daily.csv" : np.arange(0,6), "Japan_MOM_Factor_Daily.csv": np.arange(0,6), "Asia_Pacific_ex_Japan_MOM_Factor_Daily.csv": np.arange(0,6)}
data1 = {}
for keys in dic:
    data1[keys] = pd.read_csv(r"Where you left the file"+keys, skiprows = dic[keys], parse_dates = True)
    data1[keys] = data1[keys].rename(columns = {"Unnamed: 0": "TimePeriod", "WML" : "Mom"}) #renaming and debugging columns name
    data1[keys]["TimePeriod"] = pd.to_datetime(data1[keys]["TimePeriod"], format="%Y%m%d") #convert Time periods to timestamps
```

```python
for keys in data1:
#Check for missing data as described in documentation
    print(len(data1[keys][data1[keys]["Mom"] == -99.99]))
    print(len(data1[keys][data1[keys]["Mom"] == -999]))
```

```python
df1 = {}
for keys in data1:
    df1[keys] = modelling(realized_variance_test(data1[keys]), data1[keys])
```

Using the same annualized volatility of 12%, we find that the scaling factor does not reach values as high as four for other geographies than the US, which means that momentum is subject to a higher volatility in general. The other interesting fact is that except for the dot com bubble burst and the great financial crisis, geographies exhibit some idiosyncratic risks regarding the momentum strategy. Indeed, for Asia Pacific Ex-Japan, we see a burst in variance (and consequently standard deviation) around 1997 which was the Asian financial crisis whereas other geographies remain relatively calmer. We can make the same interpretation during the European sovereign crisis in 2012.

Furthermore, for all geographies, the scaled momentum approach outperforms the vanilla momentum strategy. In fact, we can reach the same conclusion than for the US regarding returns distribution which are subject to much less heavy tail. The maximum drawdowns are also reduced.

Nonetheless, for Japan any momentum strategy performed poorly. Even the scaled version offered only a Sharpe ratio of 0.17. Hence, this allows us to acknowledge that even though scaled momentum clearly outperforms its vanilla version, momentum does not have to be the go to strategy. In fact, it is assumed that this factor is the one offering the best returns prospects. Nonetheless, this assumption is usually built on samples mainly concentrated in the US stock markets.

To conclude, after having reached the same conclusion than the paper, we are now armed with a relatively easy to implement scaled momentum that outperforms its vanilla version in all samples. Nonetheless, one should think critically of this factor, as it is not the one generating the most returns in all geographies.
